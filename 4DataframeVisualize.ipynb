# Visualization of data

*   Learning Matplotlib
*   Subplots

```python
import pandas as pd
#import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C.csv", index_col=0, parse_dates=True)
```

```python
data.head()
```

```python
data.plot() # not intuiative. to add some parameters
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
data.plot()
```

```python
fig, ax = plt.subplots()
data['Close'].plot(ax=ax)
ax.set_ylabel("Price")
ax.set_title("C - Citicorp")
```

```python
fig, ax = plt.subplots(2,2)
data['Open'].plot(ax=ax[0,0], title="Open")
data['High'].plot(ax=ax[0,1], title="High")
data['Low'].plot(ax=ax[1,0], title="Low")
data['Close'].plot(ax=ax[1,1], title="Close")
plt.tight_layout()
```

```python
fig, ax = plt.subplots()
data['Volume'].loc['2020-07-01':'2020-08-30'].plot.bar(ax=ax)
```
